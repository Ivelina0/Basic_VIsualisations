## Histograms, KDE plots Rug Plots

```python
automobile_data = pd.read_csv('Automobile_data.csv', na_values = '?')

automobile_data.head(5)
```

```python
automobile_data.shape
```

```python
automobile_data.isnull().sum()

# We can see that the quality of the data is quite bad- it's not very clean
```

```python
automobile_data.dropna(inplace=True)

automobile_data.shape

# Drop all the missing values 
```

```python
plt.figure(figsize=(12,8))

sns.distplot(automobile_data['price'], color='red')

plt.title('Automobile Data')

plt.show()

# In addition to the histogram representation, the seaborn distribution plot also plots the KDE curve- probability density function 
```

```python
plt.figure(figsize=(12,8))

# Specifying the number of bins/buckets - the price of the automobile should be split into 20 buckets/bins
sns.distplot(automobile_data['price'],bins=20, color='red')

plt.title('Automobile Data')

plt.show()
```

```python
plt.figure(figsize=(12,8))

# if we only want the KDE curve but not the historgram itself
sns.distplot(automobile_data['price'],hist=False, color='blue')

plt.title('Automobile Data')

plt.show()
```

```python
plt.figure(figsize=(12,8))

# if we only want the KDE curve and the actual data points, sample that exists in the data
# Rug plot on the graph - represents individual data points in the form of vertical lines
# most closely located in the peak of the graoh
sns.distplot(automobile_data['price'],hist=False, rug=True,color='blue')

plt.title('Automobile Data')

plt.show()
```

```python
plt.figure(figsize=(12,8))

# if we only want the rug plot
sns.rugplot(automobile_data['price'],height=0.5,color='blue')

plt.title('Automobile Data')

plt.show()
```

```python
plt.figure(figsize=(12,8))

# if we only want the KDE plot no histogram
sns.kdeplot(automobile_data['price'],shade=True,color='blue')

plt.title('Automobile Data')

plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.scatterplot(x='horsepower',y='price',data=automobile_data, s=120)

plt.title('Automobile Data')

plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.scatterplot(x='horsepower',y='price',data=automobile_data,
                hue='num-of-cylinders',s=120)

plt.title('Automobile Data')

plt.show()
```

```python
sns.regplot(x='horsepower',y='price',data=automobile_data)

plt.show()

# Same scatter plot but with regression analysis and a fitted straight line with confidence intervals
```

```python
sns.regplot(x='highway-mpg',y='price',data=automobile_data)

plt.show()

# The price of a car vs the highway miles oer galon has a negative relationship
# Because higher horse power cars, tend to have lower milage and are more expensive
```

```python
sns.jointplot(x='horsepower',y='price',data=automobile_data)

plt.show()

# Univariate and bivariate relationships along the X and Y axis

# We get frequency distributions 
```

```python
sns.jointplot(x='horsepower',y='price',data=automobile_data, kind='reg')

plt.show()

# Joint plot- plus regression analysis 
# Kde and historgam
```

```python
sns.jointplot(x='horsepower',y='price',data=automobile_data, kind='kde')

plt.show()

# Instead of a scatter plot , we get a density distribution of the relationship 
```

```python
sns.jointplot(x='horsepower',y='price',data=automobile_data, kind='hex')

plt.show()
```

```python
f, ax = plt.subplots(figsize=(12,8))

sns.kdeplot(automobile_data['horsepower'],automobile_data['price'], ax=ax)

sns.rugplot(automobile_data['horsepower'], color='limegreen', ax=ax)
sns.rugplot(automobile_data['price'], color='red',vertical=True, ax=ax)

plt.title('Automobile Data')

plt.show()

# In a sngle plot, we have kde plot and rugplots layered together
```

## Visualising continuous and categorical data using different plots in Seaborn 

```python
import datetime
```

```python
tips_data = pd.read_csv('tips.csv')

tips_data.sample(10)
```

```python
tips_data.shape
```

```python
tips_data.describe()
```

```python
tips_data.groupby(by='time').mean()
```

```python
plt.figure(figsize=(12,8))

sns.barplot(x='time', y='total_bill', data=tips_data)

plt.title('Bill and Tips')
plt.xticks(rotation=90)

plt.show()
```

```python
mean_tip = tips_data['tip'].mean()

mean_tip
```

```python
tips_data['above average'] = ( tips_data['tip'] - mean_tip ) > 0

tips_data[['tip', 'above average']].head(5)
```

```python
plt.figure(figsize=(12,8))

sns.countplot('time',hue='above average', data=tips_data, order = tips_data['time'].value_counts().index)

plt.title('Bill and Tips')
#plt.xticks(rotation=90)

plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.countplot('time',hue='sex', data=tips_data, order = tips_data['time'].value_counts().index)

plt.title('Bill and Tips')
#plt.xticks(rotation=90)

plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.boxplot(x='sex', y='total_bill', data=tips_data, palette='nipy_spectral')

plt.title('Bill and Tips')

plt.show()
```

```python
plt.figure(figsize=(12,8))

sns.boxplot(x='size', y='total_bill', data=tips_data, palette='nipy_spectral')

plt.title('Bill and Tips')

plt.show()

# Clear linear relationship between party size and total bill 
```

```python
#  since the size of the party is a categorical value we can visualise it like this

plt.figure(figsize=(12,8))

sns.catplot(x='size', y='total_bill',hue='time', data=tips_data,kind='bar', height=6, aspect=2, palette='CMRmap')

plt.title('Bill and Tips')

plt.show()

# Bar plots for total bill 
```

```python
plt.figure(figsize=(12,8))

sns.stripplot(x='size', y='tip',data=tips_data)

plt.title('Bill and Tips')

plt.show()

# Distribution of data points for a categorial variable, using the strip plot is great 
# it is a scatterplot representation for categprocal variables 
# in a strip plot all the values are clustered together for a particular category
```

```python
# if you want to view them in a spread out manner so that individual data points are visiblue then we can use the following
# scatter plot for individual values for categorical data

plt.figure(figsize=(12,8))

sns.swarmplot(x='size', y='tip',hue='above average', data=tips_data)

plt.title('Bill and Tips')

plt.show()
```

```python
#seperating the hue values 
plt.figure(figsize=(12,8))

sns.swarmplot(x='size', y='tip',hue='above average', data=tips_data, dodge=True)

plt.title('Bill and Tips')

plt.show()
```

## Visualizing and customising pairwise relationships using the PairGrid

```python
# using the old data 
automobile_data = pd.read_csv('Automobile_data.csv', na_values = '?')

automobile_data.dropna(inplace=True)
automobile_data.columns
```

```python
# using this to simplify our next visualisation

automobile_subset = automobile_data[['horsepower', 'city-mpg', 'highway-mpg', 'price']]
```

```python
sns.pairplot(automobile_subset)

plt.show()
```

```python
# or using the vars to only veiew relationhips i have specified

sns.pairplot(automobile_data, vars=['price','horsepower','highway-mpg'])

plt.show()
```

```python
sns.pairplot(automobile_data, vars=['price','horsepower','highway-mpg'], hue='fuel-type')

plt.show()

# all of the relationships will be expressed based on the catogries of whether its 
# a diesel powerder car or petrol powered car
# Our univariate data has now been expressed using a KDE curves, two curves for fuel type: gas and fuel type diesel
#  the scatterplots also have the same split- datapoints are colored based on the fuel type
```

```python
g = sns.PairGrid(automobile_data, vars=['price','horsepower','highway-mpg'],hue='fuel-type')

g.map(plt.scatter)

g.add_legend()

plt.show()

# The same information is displayed
# The map specifies that all plots should be scatter plots 
# The major change is shown along the diagonal line - its a stright line
```

```python
g = sns.PairGrid(automobile_data, vars=['price','horsepower','highway-mpg'],hue='fuel-type')

g.map_diag(plt.hist)
g.map_offdiag(plt.scatter)

g.add_legend()

plt.show()

# the same data but here the main diagonal is set as histograms and the off diagonal to be scatterplots
# Two histograms layed on top of each other
```

```python
g = sns.PairGrid(automobile_data, vars=['price','horsepower','highway-mpg'],hue='fuel-type')

g.map_lower(plt.scatter)
g.map_diag(sns.kdeplot, lw=3, legend=False)
g.map_upper(sns.regplot)

g.add_legend()

plt.show()

# this time the lower portion of the pairgrid to show scatterplots
# the diagonal bit of the grid shows the kde curves by fuel type
# and the upper portion of the pairwise relationship is represented using regression plots
```

```python
automobile_corr = automobile_data[['engine-size', 'horsepower','peak-rpm','city-mpg','highway-mpg','price']].corr()

automobile_corr
# subset of variables i want to correlate using a heatmap
# first we find the correlations using .corr()
```

```python
plt.figure(figsize=(12,8))

sns.heatmap(automobile_corr, vmax=.8, square=True, annot=True,fmt='.2f', cmap='viridis' )

plt.title('Automobile Correlations')

plt.show()
```

## Visualising multiple Relationships using facets

```python
plt.figure(figsize=(15,10))

fg = sns.FacetGrid(automobile_data, col='fuel-type', height=7, aspect=1)

fg.map(plt.hist, 'price', color='red')

plt.show()

# here we want to see how the price of a car is distributed based on the fuel type
# The hist views the price distribution
# Two hist in two seperate columns,
# Fuel type has two categpries and these are the two facets of the same relationship
```

```python
# Here we visualise the same data but we are using the facetgrid to visualise a bivariate relationship, compared to the univariate abpve
# Set up a facegrid: column represent the fuel type
# we use a scatterplot to see how the milage of a car avries based on price, for two facets- fuel type gas and diesel

plt.figure(figsize=(15,10))

fg = sns.FacetGrid(automobile_data, col='fuel-type', height=7, aspect=1)

fg.map(sns.scatterplot,'highway-mpg', 'price', s=100)

fg.add_legend()

plt.show()
```

```python
# In addition to having the two columns that indicate the fuel type, we also indicate the hue of the scatterplot 
# to see the number of cylinders a car has
# Cars with 6 cylinders tend to have low milage and high price

plt.figure(figsize=(15,10))

fg = sns.FacetGrid(automobile_data, col='fuel-type',hue='num-of-cylinders', height=7, aspect=1)

fg.map(sns.scatterplot,'highway-mpg', 'price', s=100)

fg.add_legend()

plt.show()
```

```python
## using the tipsdata

# the columns are the size of the party and seeing how the total bill varies based on the different sized parties

plt.figure(figsize=(15,10))

fg = sns.FacetGrid(tips_data, col='size', height=7, aspect=1)

fg.map(plt.hist,'total_bill', color= 'green')

#fg.add_legend()

plt.show()

# we have 6 columns, and most of our data is for the party size 2
# i.e most of the people dine in pairs
```
